```python
from os import listdir
from os.path import isfile, join
import numpy as np
import pandas as pd
from IPython.display import display
import datetime
import re
import csv
from geopy.geocoders import Nominatim

table = pd.read_csv('inputs/list_of_corps.csv').set_index('Corps')
display(table)

geolocator = Nominatim(user_agent="hi")
for corps in table.index:
    print(corps)
    slice = table.loc[corps, "Home Location"]
    location = geolocator.geocode(slice)
    table.loc[corps, "Latitude"] = location.latitude
    table.loc[corps, "Longitude"] = location.longitude

table.to_csv(f'exports/with_locations.csv')
```
